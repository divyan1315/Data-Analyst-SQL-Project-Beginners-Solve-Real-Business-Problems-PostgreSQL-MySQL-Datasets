```python
# =====================================
# SAMPLE SUPERSTORE - COMPLETE WORKFLOW
# =====================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import zipfile
import os

# ------------------------------
# 1. Check Files
# ------------------------------
print("Files before extraction:")
print(os.listdir())

# ------------------------------
# 2. Extract ZIP file
# ------------------------------
zip_path = "archive (8).zip"   # change if needed

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall()

print("\nFiles after extraction:")
print(os.listdir())

# ------------------------------
# 3. Find CSV Automatically
# ------------------------------
# After extraction, we know the name of the CSV file from the zip archive.
csv_file = "Sample - Superstore.csv"

print("\nCSV Found at:", csv_file)

# ------------------------------
# 4. Load Dataset
# ------------------------------
df = pd.read_csv(csv_file, encoding="latin1")

print("\nDataset Loaded Successfully!\n")
print(df.head())

# ------------------------------
# 5. Convert Dates
# ------------------------------
df["Order Date"] = pd.to_datetime(df["Order Date"], errors="coerce")
df["Ship Date"] = pd.to_datetime(df["Ship Date"], errors="coerce")

# ------------------------------
# 6. Missing Values
# ------------------------------
print("\nMissing Values:")
print(df.isnull().sum())

# ------------------------------
# 7. Basic Analysis
# ------------------------------
print("\nTotal Sales:", df["Sales"].sum())
print("Total Profit:", df["Profit"].sum())

# ------------------------------
# 8. Sales by Category
# ------------------------------
category_sales = df.groupby("Category")["Sales"].sum()

plt.figure()
category_sales.plot(kind="bar")
plt.title("Sales by Category")
plt.ylabel("Sales")
plt.show()

# ------------------------------
# 9. Profit by Region
# ------------------------------
region_profit = df.groupby("Region")["Profit"].sum()

plt.figure()
region_profit.plot(kind="bar")
plt.title("Profit by Region")
plt.ylabel("Profit")
plt.show()

# ------------------------------
# 10. Monthly Sales Trend
# ------------------------------
df["Month"] = df["Order Date"].dt.to_period("M")
monthly_sales = df.groupby("Month")["Sales"].sum()

plt.figure()
monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.xticks(rotation=45)
plt.show()

# ------------------------------
# 11. Correlation Heatmap
# ------------------------------
plt.figure()
sns.heatmap(df[["Sales","Profit","Discount","Quantity"]].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()

print("\n✅ PROJECT ANALYSIS COMPLETED")
```
